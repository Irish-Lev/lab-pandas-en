### 4 - Now return all the rows where the CGPA is greater than 9 and the SOP score is less than 3.5. Find the mean chance of admit for these applicants.

```python
# your code here
# Firstly, filter the DataFrame based on the conditions and then count the number of rows in the filtered DataFrame.
filtered = admissions[(admissions["CGPA"] > 9) & (admissions["SOP"] < 3.5)]

# Calculating the 'mean' of the 'Chance of Admit' column for the filtered DataFrame
mean_chance_of_admit = filtered["Chance of Admit "].mean()
mean_chance_of_admit
```

```python
# Alternative solution
mean_chance = admissions[(admissions["CGPA"] > 9) & (admissions["SOP"] < 3.5)]["Chance of Admit "].mean()
mean_chance
```
